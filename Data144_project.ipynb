```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data = pd.read_csv("/content/drive/My Drive/DMA project/data/weather_data.csv")
```

```python
#create column "Wildfire_Occurred"(set as below criteria) 0 or 1
data["Wildfire_Occurred"] = data["Wildfire_Occurred"] = (
    (data["Precipitation_mm"] <= 5) &
    (data["Wind_Speed_kmh"] >= 15)
).astype(int)
data['Month'] = pd.to_datetime(data["Date_Time"]).dt.month    # add this for predict wildfire risk location and month
```

```python
data
```

```python
#set x(features) and y(target)
x = data.drop(columns=["Wildfire_Occurred", "Date_Time", "Location"])
y = data["Wildfire_Occurred"]
```

```python
#split train test set
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
#ML Random Forest
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=10,
    random_state=42
)
rf_model.fit(X_train, y_train)
rf_predict = rf_model.predict(X_val)
```

```python
#Accuracy
accuracy = accuracy_score(y_val, rf_predict)
print(f"Accuracy: {accuracy:.2f}")
```

```python
#predict with value on data
first_predict = pd.DataFrame(
    [[38.304456, 32.298644, 2.570483, 22.996614	, 4]],
    columns = x.columns
)
first_prediction = rf_model.predict(first_predict)
print(f"Predicted Class: {first_prediction}")
```
